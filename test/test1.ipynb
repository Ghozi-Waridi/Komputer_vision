```python
CLASS_NAMES = ['glioma', 'meningioma', 'notumor', 'pituitary']
CLASS_TO_INDEX = {name: idx for idx, name in enumerate(CLASS_NAMES)}

all_labels = []
for class_name in CLASS_NAMES:
    all_labels.extend([CLASS_TO_INDEX[class_name]] * 1)   
```
